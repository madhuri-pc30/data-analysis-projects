```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("fitness_data.csv")
df.head()
```

```python
plt.plot(df["steps"])
plt.title("Raw Steps Data")
plt.ylabel("Steps")
plt.xlabel("Row of data")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df["date"], df["steps"])
plt.title("Number of Steps per Day")
plt.ylabel("Number of Steps")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["date"],df["resting_hr"],color ="darkred", linewidth =2)
plt.title("Resting Heart Rate Over Time", fontsize =25)
plt.ylabel("Resting Heart Rate (bpm)")
plt.xlabel("Date", fontsize =20)
plt.tight_layout()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df["steps"], df["calories_burned"], alpha=0.6, color="slateblue")
plt.title("Steps vs Calories Burned", fontsize=15)
plt.xlabel("Steps", fontsize=10)
plt.ylabel("Calories Burned", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,6))
colors= {"Low":"lightblue", "Moderate":"green", "High":"Orange"}
for level ,subset df.groupby("activity_level"):
    plt.scatter(subset["steps"], subset["calories_burned"], alpha=0.6, label=level, color=colors[level])
```

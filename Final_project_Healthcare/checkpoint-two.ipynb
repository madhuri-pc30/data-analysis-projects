## Visualize

```python
# 1. Medical Condition Counts
# This bar chart shows which medical conditions appear the most.
# It helps me understand what types of health issues are common in this dataset.
condition_counts = health_df['Medical Condition'].value_counts()
plt.bar(condition_counts.index, condition_counts.values)
plt.title("Medical Condition Counts")
plt.xlabel("Medical Condition")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
